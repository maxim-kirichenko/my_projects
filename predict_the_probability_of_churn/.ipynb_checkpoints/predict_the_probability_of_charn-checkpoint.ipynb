#### Описание проекта:

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с пользователями на основе аналитических данных.
Необходимо провести анализ имеющихся данных и подготовить план действий по удержанию клиентов.

#### Описание данных:

- 'Churn' — факт оттока в текущем месяце;
- 'gender' — пол
- 'Near_Location' — проживание или работа в районе, где находится фитнес-центр
- 'Partner' — сотрудник компании-партнёра клуба
- Promo_friends — факт первоначальной записи в рамках акции «приведи друга»
- 'Phone' — наличие контактного телефона
- 'Age' — возраст
- 'Lifetime' — время с момента первого обращения в фитнес-центр (в месяцах)
- 'Contract_period' — длительность текущего действующего абонемента (месяц, 3 месяца, 6 месяцев, год)
- 'Month_to_end_contract' — срок до окончания текущего действующего абонемента (в месяцах)
- 'Group_visits' — факт посещения групповых занятий
- 'Avg_class_frequency_total' — средняя частота посещений в неделю за все время с начала действия абонемента
- 'Avg_class_frequency_current_month' — средняя частота посещений в неделю за предыдущий месяц
- 'Avg_additional_charges_total' — суммарная выручка от других услуг фитнес-центра: кафе, спорт-товары, косметический и массажный салон

## Шаг 3. Постройте модель прогнозирования оттока пользователей. <a id="step7"></a>

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = df.drop(columns = ['churn'])
y = df['churn']

# разделяем модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
def regression_model(models):
    model = models
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    print("accuracy:", accuracy)
    print("precision:", precision)
    print("recall:", recall)
```

```python
regression_model(LogisticRegression(random_state=0))
```

```python
regression_model(RandomForestClassifier(random_state=0))
```

Модель "Случайный лес" показывает немного меньшую долю правильных ответов, чем мем модель "логистическая регрессия". Но в целом в обоих случаях доля правильных ответом стремится к 1.
Модель логистрической регресси показывает болеее точные результаты.

## Шаг 4. Сделайте кластеризацию пользователей. <a id="step8"></a>

### 4.1. Стандартизируйте данные.<a id="step9"></a>


```python
#Отложим в сторону столбец с оттоком
X = df.drop(columns = ['churn'])
```

```python
scaler = StandardScaler()
X_sc = scaler.fit_transform(X)
```

### 4.2. Постройте матрицу расстояний функцией linkage() на стандартизованной матрице признаков и нарисуйте дендрограмму.  На основании полученного графика предположите, какое количество кластеров можно выделить. <a id="step10"></a>


```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linked = linkage(X_sc, method = 'ward')
plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.show()
```

#### Из полученной выше дендограммы выделяется 4 кластеров.

### 4.3. Обучите модель кластеризации на основании алгоритма K-Means и спрогнозируйте кластеры клиентов. Договоримся за число кластеров принять n=5, чтобы ваши результаты можно было сравнивать с результатами остальных студентов. Однако, конечно, в жизни никто не скажет вам правильный ответ, и решение остаётся за вами (на основании изучения графика из предыдущего пункта). <a id="step11"></a>

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters = 5, random_state=1) # задаём число кластеров, равное 5
labels = km.fit_predict(X_sc) # применяем алгоритм к данным и формируем вектор кластеров
```

```python
labels
```

	Доля уходящих в отток велика среди проживающих в других районах. Доля оттока меньше серди сотруднтиков компаний-партнеров.
	Те, кто не посещают групполвые занятие, чаще уходят в отток.

Большая корреляция между средним посещением в неделю за предыдущий месяц и средней частотой посещений в неделю за все время с начала действия абонемента. Также большая корреляция между количеством месяцев до конца текущего абонемента и сроком действия абонемента.


<H3> Харатеристика кластеров аудитории: </H3><a id="step18"></a>

<b> Кластер 0: </b>

В кластер попали пользователи, которые проживают в других районах (не в том, в котором находится зал). Лишь 7% из них пришли по акции "Приведи друга". У всех указан номер телефона. У данной группы самая высоккая вероятность оттока.

<b>Кластер 1:</b>

Пользователи проживают в районе, в котором находится фитнес клуб. Ни один из пользователей не пришел по акции "приведи друга". У всех есть номера телефонов.

<b>Кластер 2:</b> Практически все пользователи проживают в этом районе. Большинство из посетителей являются сотрудниками компаний-партнеров. Практически половина пришла по акции "приведи друга". У всех есть номера телефонов. Группа характеризуется высокой средней продолжительностью действующего контракта, а также самой высокой посещаемостью групплвых занятий. У группы самые высокие средние затраты на дополнительные услуги. У группы самое высокое количество месяцев до окончания абонемента, вероятно связано с тем, что у этой группы в целом самые "длинные" абонементы. У группы самая высокая средняя посещаемость, как месячная посещаемость. У группы самый низкий показатель оттока.

<b>Кластер 3:</b> Характеризуется проживаением в районе нахождения клуба. Все пользователи пришли по акции "Приведи друга". У всех есть номера телефонов.

<b>Кластер 4:</b> Самая яркая отличительная черта - нет номеров телефонов. Достаточго высокая средняя продолжительность контракта - 4.7 месяцев.


<h3>Выводы:</h3>
Самая высокая вероятность оттока у посетителей из кластера 0. Самая низкая - у посетителей из кластера 2

<h3>Рекомендации:</h3><a id="step19"></a>

- Пользователи, которые приобретабт абоненмент на 1 месяц наиболее склонны к отттоку, можно попробовать предллагать им второй месяц в подарок или за пол цены.
- Также необходимо мотивировать посетителей на более састое посещение зала в течении недели. Эьто могут быть как напоминания и приглашения, так и какие-то специапльные активности.
- Требуется дополнительная мотивайия для клиентов, которые живут в других районах. 
- Нужно получить телефон клиента. Также важно посещение групповых занятий.
- Привлекать больше клиентов по акции "Приведи друга"
