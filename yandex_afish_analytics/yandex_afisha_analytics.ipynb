#### Описание проекта:

В распоряжении есть данные от Яндекс.Афиши с июня 2017 по конец мая 2018 года:
- лог сервера с данными о посещениях сайта Яндекс.Афиши,
- выгрузка всех заказов за этот период,
- статистика рекламных расходов.

Изучить: 
- как люди пользуются продуктом,
- когда они начинают покупать,
- сколько денег приносит каждый клиент
- когда клиент окупается.

#### Описание данных:

##### Таблица visits (лог сервера с информацией о посещениях сайта):
- Uid — уникальный идентификатор пользователя
- Device — категория устройства пользователя
- Start Ts — дата и время начала сессии
- End Ts — дата и время окончания сессии
- Source Id — идентификатор рекламного источника, из которого пришел пользователь

##### Таблица orders (информация о заказах):
- Uid — уникальный id пользователя, который сделал заказ
- Buy Ts — дата и время заказа
- Revenue — выручка Яндекс.Афиши с этого заказа
##### Таблица costs (информация о затратах на маркетинг):
- source_id — идентификатор рекламного источника
- dt — дата
- costs — затраты на этот рекламный источник в этот день

Нормальные значения распределны в диапазоне от 100 до 900. <br>
Избавимся от выбросов

```python
def remove_outlier(data, col):
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1 
    fence_low  = q1 - 1.5 * iqr
    fence_high = q3 + 1.5 * iqr
    df_out = data.loc[(data[col] > fence_low) & (data[col] < fence_high)]
    return df_out
df_visits_clean = remove_outlier(df_visits, 'session_duration_sec')
```

```python
df_visits_clean['session_duration_sec'].describe()
```

После очистки от выбросов, получилось, что средняя продолжительность сессии составляет 438 секунд, или 7,3 минуты.

Посчитаем retention rate

Так как данные в датафрейме представлены за несколько лет, то для удобства их анализа сделаем столбец, где отобразим месяц и год начала сессии, как первое число соотвествующего года и месяца:

```python
df_visits['session_year_month'] = df_visits['session_date'].values.astype('datetime64[M]')
```

Возьмём дату, когда пользователь впервые проявил активность в мобильном приложении. И объединим его с исходным датафреймом.

```python
first_activity_date = df_visits.groupby(['uid'])['session_date'].min()
```

Переименуем объект Series, так как в исходном датафрейме уже есть столбец session_date , назовем его first_activity_date

```python
first_activity_date.name = 'first_activity_date'
df_visits = df_visits.join(first_activity_date, on='uid')
```

Так как большинство пользователей совершают только одну сессию, считаю, что достаточно будет посмотроит когорты по месяцам. Определим когорты по месяцам, для этого запишем first_activity_date как первое число месяца и года совершения сессии.

```python
df_visits['first_activity_month'] = df_visits['first_activity_date'].astype('datetime64[M]')
df_visits.head()
```

Cгруппируем данные по когорте first_activity_month и месяцу сессии session_year_month 

```python
sessions_grouped_by_cohorts = df_visits.groupby(['first_activity_month', 'session_year_month']).agg({'uid':'nunique'}).reset_index()
```

Построим сводную таблицу изменения кол-ва уникальных пользователей в когортах по месяцу сессии и оценим, как изменяется их количество с течением времени

```python
sessions_grouped_by_cohorts.pivot_table(index='first_activity_month',columns='session_year_month',values='uid',aggfunc='mean')
```

Найдём lifetime когорты. Из месяца каждой сессии для уникального пользователя вычтем месяц первой сессии.

```python
df_visits['months_from_firs_session'] = df_visits['session_year_month'] - df_visits['first_activity_month']
```

Получили столбец с типом данных timedelta, показывающий количество дней между датами. Преобразуем этот тип в число месяцев, прошедших между датами. Для этого months_from_firs_session нужно разделить на константу np.timedelta64(1, 'M')(есть только в библиотеке numpy) — узнать, сколько величин длиной в 1 месяц содержится в cohort_lifetime. 


```python
df_visits['months_from_firs_session'] = df_visits['months_from_firs_session'] / np.timedelta64(1,'M')
```

```python
df_visits['months_from_firs_session'] = df_visits['months_from_firs_session'].round().astype('int')
```

оставим в обозначении когорты только год и месяц. 

```python
df_visits['first_activity_month'] = df_visits['first_activity_month'].dt.strftime('%Y-%m')
```

```python
sessions_grouped_by_cohorts = df_visits.groupby(['first_activity_month', 'months_from_firs_session']).agg({'uid':'nunique'}).reset_index()
```

Чтобы найти Retention Rate, нужно сперва получить число пользователей, изначально бывших в когорте, и на него разделить число пользователей в каждую следующую неделю.
Найдём исходное количество пользователей в когорте. Возьмём их число на нулевую неделю и объединим данные по когортам с исходным количеством пользователей в когорте:


```python
initial_users_count = sessions_grouped_by_cohorts[sessions_grouped_by_cohorts['months_from_firs_session'] == 0][['first_activity_month','uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})

initial_users_count.head(10)
```

```python
sessions_grouped_by_cohorts = sessions_grouped_by_cohorts.merge(initial_users_count, on='first_activity_month')
```

```python
sessions_grouped_by_cohorts['retention'] = sessions_grouped_by_cohorts['uid'] / sessions_grouped_by_cohorts['cohort_users']
sessions_grouped_by_cohorts.head(10)
```

```python
retention_pivot = sessions_grouped_by_cohorts.pivot_table(index='first_activity_month', columns='months_from_firs_session', values='retention',aggfunc='mean')
sns.set(style='white')
plt.figure(figsize=(20, 9))
plt.title('Cohorts: User Retention')
plt.yticks(rotation=90)
sns.heatmap(retention_pivot, vmin=0, vmax=1, center=0.5, annot=True, fmt='.1%', linewidths=1, linecolor='gray');
```

Очень большой отток пользователей после первого месяца. во второй месяц возвращается только 5-8% пользователей.

```python
cohorts = orders.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
cohorts.head()
```

```python
cohort_sizes = orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohorts = cohorts.merge(cohort_sizes, on='first_order_month')
cohorts
```

```python
cohorts['ltv'] = cohorts['revenue'] / cohorts['n_buyers']

ltv = cohorts.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean')

ltv.cumsum(axis=1).round(2)
```

```python
paying_users_1_session = (df_visits[df_visits['uid'].isin(df_orders['uid'].unique())]
                      .sort_values(by = ['uid', 'start_ts']).groupby(['uid', ])
                      .agg({'start_ts':'min', 'source_id': 'first', 'device': 'first'}).rename(columns={'start_ts':'acquisition_date'})).reset_index()
paying_users_all_sessions = (df_visits[df_visits['uid'].isin(df_orders['uid'].unique())]
                              .sort_values(by = ['uid', 'start_ts']))[['uid', 'start_ts']]
                             
paying_users_all_sessions1 = paying_users_all_sessions.merge(paying_users_1_session, left_on='uid', right_on='uid')

paying_users_all_sessions1['acquisition_ymd'] = paying_users_all_sessions1['acquisition_date'].dt.date
paying_users_all_sessions1['acquisition_month'] = paying_users_all_sessions1['acquisition_date'].astype('datetime64[M]') 
paying_users_1_session = paying_users_all_sessions1.groupby('uid').agg({'source_id':'first', 'device':'first'}).reset_index()
paying_users_1_session
```

Посчитаем romi всех пдлатящих клиентов:

```python
# в таблице с заказами отберем тех пользоватлей, которые приходили с 1-ого источника
source_1_uid = paying_users_1_session

# из df_orders  оставим толькко тех пользователей, которые пришли из первого источника, т.е. тех, чью uid присутствуют в переменной source_1_uid
df_orders['order_month'] = df_orders['transaction_date'].astype('datetime64[M]')
source_1_orders = df_orders[df_orders['uid'].isin(source_1_uid['uid'])]

#из source_1_orders получим месяц первой покупки каждого покупателя.
source_first_orders = source_1_orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
source_first_orders.columns = ['uid', 'first_order_month']

#сгруппируем по месяцу первой покупки и посчитаем кколичество уникальных uid в этом месяце
source_cohort_sizes = source_first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
source_cohort_sizes.columns = ['first_order_month', 'n_buyers']

#Добавим месяц первой покупки каждого покупателя в таблицу с заказами source_1_orders
source_orders_new = pd.merge(source_1_orders,source_first_orders, on='uid')

#группируем получившуюся таблицу по месяцу первой покупки и месяцу каждого заказа. Выручку суммируем
source_cohorts = source_orders_new.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
source_cohorts

#добавим количество людей в каждой когорте
source_report = pd.merge(source_cohort_sizes, source_cohorts, on='first_order_month')
source_report

#Определим возраст когорты
source_report['age'] = (source_report['order_month'] - source_report['first_order_month']) / np.timedelta64(1, 'M')
source_report['age'] = source_report['age'].round().astype('int')

#найдем LTV
source_report['ltv'] = source_report['revenue'] / source_report['n_buyers']


#считаем САС. Для начала оставим только расходы на искомый источник трафика
#source_df_costs = df_costs[df_costs['source_id'] == 1]

#Найдем суммарные расходы за каждый месяц
source_monthly_costs = df_costs.groupby('month')['costs'].sum().to_frame()

#добавим данные из расходов в таблицу с ltv и посчитаем САС
source_report_new = pd.merge(source_report, source_monthly_costs, left_on='first_order_month', right_on='month')
source_report_new['cac'] = source_report_new['costs'] / source_report_new['n_buyers']

#считаем ROMI
source_report_new['romi'] = source_report_new['ltv'] / source_report_new['cac']
source_output = source_report_new.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='romi', 
        aggfunc='mean')
 
source_output.cumsum(axis=1).round(2)
```
