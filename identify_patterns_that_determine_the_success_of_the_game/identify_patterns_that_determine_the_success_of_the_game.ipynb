##### Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и постройем распределение по годам. Ответим на вопрос: за какой характерный срок появляются новые и исчезают старые платформы?<a id="step4"></a>

Еще раз выберем список платформ, кол-во строк и сумму продаж и средний чек:

```python
platform_sales = df.groupby('platform')['total_sales'].agg(['count','sum']).sort_values('sum', ascending=False)
platform_sales['average_check'] = platform_sales['sum'] / platform_sales['count']
platform_sales['platform'] = platform_sales.index
platform_sales
```

Выведем таблицу с первым годом появления игры и последним годом появления игры, рассчитаем разницу между этими годами и объединими таблицы по индексу:

```python
platform_years = df.groupby('platform')['year_of_release'].agg(['min','max'])
platform_years['life_of_platform'] = platform_years['max'] - platform_years['min']
platform_years
```

```python
platform_sales_pivot = platform_sales.join(platform_years)
platform_sales_pivot['av_annual_income'] = platform_sales_pivot['sum'] / platform_sales_pivot['life_of_platform']
platform_sales_pivot.columns = ['games_count', 'total_income', 'average_check', 'platform', 'first_year', 'last_year', 'life_of_platform', 'av_annual_income']
platform_sales_pivot.reset_index(drop=True, inplace=True)
platform_sales_pivot
```

Абсолютным лидером по продажам является платформа ps2 , на которую было выпущено больще всего игр. Практически в 2 раза больше, чем на x360. В среднем одна игра приносила 0.58 млн. долларов. Платформа просуществовала 11 лет, с 2000 по 2011 года. 

Выведем на графике суммарный доход по платформам и среднегодовой доход по платформе;

```python
ax = platform_sales_pivot.plot(x='platform', y='total_income', figsize=(20, 10))
platform_sales_pivot.plot(x='platform', y='av_annual_income', figsize=(20, 10), ax=ax)
```

На графике виден резкий переход между ps (727 млн) и ps4 (314 млн), однако данные по ps4 приведеные за 3 года (с 2013 по 2016) и ps4 имеет самый высокий среднегодовой доход, сопоставимый с продажами лидера - ps2, так я бы включил эту платформу для дальнейшего анализа.

# Шаг 4. Проверьте гипотезы<a id="step14"></a>

##### Проверим гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые;<a id="step15"></a>

Создадим 2 датафрейма, каждый из которых включает только пользователей этих платформ

```python
xone_gamers = df_popular_platforms.query('platform == "xone" and (user_score != "NaN")')
pc_gamers = df_popular_platforms.query('platform == "pc"and (user_score != "NaN")')

xone_gamers = xone_gamers.reset_index()
xone_gamers = xone_gamers.drop('index', axis=1)

pc_gamers = pc_gamers.reset_index()
pc_gamers = pc_gamers.drop('index', axis=1)
```

Проверим распределение:

```python
xone_gamers.hist('user_score', bins=10)
pc_gamers.hist('user_score', bins=10)
```

```python
xone_gamers['user_score'].describe()
```

```python
pc_gamers['user_score'].describe()
```

Распределения похожи на нормальные. 
При проверке гипотезы о равенстве двух генеральных совокупностей судим по имеющимся в нашем распоряжкении выборкам. (данные за 2016 год не полные).
Распределения похожи на нормальные. 
Будем использовать t-test для проверки гипотезы.
Чтобы проверить гипотезу о равенстве среднего двух генеральных совокупностей по взятым из них выборкам, применяется метод ttest_ind. 

Нулевая гипотеза: 
Нулевая гипотеза H₀: "Средние пользовательские рейтинги одной платформы равны средним польщовательским рейтингам другой платформы";

Альтернативная гипотеза H₁: "Средние пользовательские рейтинги одной платформы не равны средним пользовательским рейтингам другой платформы"

```python
alpha = .05
xone_users = xone_gamers['user_score']
pc_users = pc_gamers['user_score']

result = st.ttest_ind(xone_users, pc_users)

print('p-значение:', result.pvalue)

if (result.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу")
```

Гипотезу о том, что средни средние пользовательские рейтинги одной платформы равны средним польщовательским рейтингам другой платформы отвергаем. 

##### Проверим гипотезу: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.<a id="step16"></a>

Нулевая гипотеза H₀: "Средние пользовательские рейтинги жанров Action и Sports равны";

Альтернативная гипотеза H₁: "Средние пользовательские рейтинги жанров Action и Sports не равны"

```python
alpha = .05
action_gamers = df_popular_platforms.query('genre == "action" and (user_score != "NaN")')['user_score']
sports_gamers = df_popular_platforms.query('genre == "sports" and (user_score != "NaN")')['user_score']

result = st.ttest_ind(action_gamers, sports_gamers)

print('p-значение:', result.pvalue)

if (result.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не отвергаем нулевую гипотезу")
```

Гипотезу о том, что  средние пользовательские рейтинги жанров Action и Sports равны отвергаем.

Как и говорилось в задании. Пользовательские рейтинги разные!
