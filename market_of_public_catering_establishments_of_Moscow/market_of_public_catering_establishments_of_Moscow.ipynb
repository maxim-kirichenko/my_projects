#### 3. Для какого вида объекта общественного питания характерно сетевое распространение? <a id="step5"></a>

```python
pivot_object_type = df.pivot_table(index='object_type',
              columns='chain',
              values='count',
              aggfunc='sum').reset_index()
pivot_object_type['sum'] = pivot_object_type['да'] + pivot_object_type['нет']
pivot_object_type['chain_part'] = pivot_object_type['да'] / pivot_object_type['sum'] *100
pivot_object_type
```

```python
sns.barplot(x='object_type', y='chain_part', data=pivot_object_type)
plt.xticks(rotation=90);
plt.title('Доля сетевых заведений')
plt.xlabel('Типы объектов')
plt.ylabel('Кол-во объектов');
```

#### 10 .Посмотрите на распределение количества посадочных мест для улиц с большим количеством объектов общественного питания. Какие закономерности можно выявить? <a id="step12"></a>

```python
popular_streets = result[result['street'].isin(top10_streets['street'].unique())]
sns.distplot(popular_streets['number'])
plt.title('Распределение кол-ва посадочных мест для улиц ТОП-10')
plt.xlabel('Кол-во посадочных мест')
plt.ylabel('Кол-во заведений');
```

```python
popular_streets['number'].describe()
```

```python
popular_streets['number'].mode()
```

Hаспределение не нормальное, скошено вправо.
- Среднее - 51
- Медина - 25
- Мода - 0

Много выбросов

```python
sns.boxplot(x="count", y='number', data=popular_streets)
```

Есть выбросы, попробуем очистить

```python
def remove_outlier(data, col):
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1 
    fence_low  = q1 - 1.5 * iqr
    fence_high = q3 + 1.5 * iqr
    df_out = data.loc[(data[col] > fence_low) & (data[col] < fence_high)]
    return df_out
popular_streets_clean = remove_outlier(popular_streets, 'number')
```

```python
popular_streets_clean['number'].describe()
```

```python
popular_streets_clean['number'].mode()
```

```python
sns.distplot(popular_streets_clean['number'])
plt.title('Распределение кол-ва посадочных мест для улиц ТОП-10')
plt.xlabel('Кол-во посадочных мест')
plt.ylabel('Кол-во заведений');
```

После очистки все равно остались выбросы. Среднее значение подверглось уменьлению, а медианга осталась практически на прежнем уровне. Распределенние не нормальное. Правильнее будет ориентироваться по медиане - среднее кол-во посадочных мест 24
